```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, classification_report
from sklearn import datasets

# Create a synthetic dataset
x, y = datasets.make_blobs(
    n_samples=150, n_features=2, centers=2, cluster_std=1.05, random_state=2
)
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=123
)

# Instantiate and train the Perceptron classifier
perceptron = Perceptron(max_iter=1000, tol=1e-3, random_state=0)
perceptron.fit(X_train, y_train)

# Make predictions
predictions = perceptron.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print("Perceptron classification accuracy:", accuracy)

# Generate a classification report
report = classification_report(y_test, predictions)
print("Classification Report:\n", report)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import classification_report

class Perceptron:
    def __init__(self, learning_rate=0.1, n_iters=10000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.activation_func = self._unit_step_func
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize parameters
        self.weights = np.zeros(n_features)
        self.bias = 0

        y_ = np.array([1 if i > 5 else 0 for i in y])

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_func(linear_output)

                # Perceptron update rule
                update = self.lr * (y_[idx] - y_predicted)

                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_func(linear_output)
        return y_predicted

    def _unit_step_func(self, x):
        return np.where(x >= 0, 1, 0)

def accuracy(y_true, y_pred):
    accuracy = np.sum(y_true == y_pred) / len(y_true)
    return accuracy

# Create a synthetic dataset
x, y = datasets.make_blobs(
    n_samples=150, n_features=2, centers=2, cluster_std=1.05, random_state=2
)
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=123
)

# Instantiate and train the perceptron with adjusted parameters
p = Perceptron(learning_rate=0.1, n_iters=10000)
p.fit(X_train, y_train)

# Make predictions
predictions = p.predict(X_test)

# Calculate accuracy
accuracy = accuracy(y_test, predictions)
print("Perceptron classification accuracy:", accuracy)

# Generate a classification report
report = classification_report(y_test, predictions)
print("Classification Report:\n", report)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.utils import resample
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming X and y are your feature and target variables, respectively.

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Define the number of bootstrap iterations
n_iterations = 100  # You can adjust this as needed.

# Initialize a list to store accuracy scores from each iteration
accuracy_scores = []

# Initialize a list to store one confusion matrix
confusion_matrices = []

# Create and train the k-NN classifier for each bootstrap iteration
k = 3  # You can choose the number of neighbors (k) as needed.
n_components = 2  # Number of principal components for PCA

for _ in range(n_iterations):
    # Resample the training data with replacement (bootstrap)
    X_resampled, y_resampled = resample(X_train, y_train, random_state=42)

    # Apply PCA to the resampled data
    pca = PCA(n_components=n_components)
    X_resampled_pca = pca.fit_transform(X_resampled)

    # Create and train the k-NN classifier
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(X_resampled_pca, y_resampled)

    # Apply PCA to the test data
    X_test_pca = pca.transform(X_test)

    # Make predictions on the test data
    y_pred = knn_classifier.predict(X_test_pca)

    # Calculate accuracy and store it in the list
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

    # Generate a confusion matrix for one iteration and store it
    if len(confusion_matrices) < 1:
        conf_matrix = confusion_matrix(y_test, y_pred)
        confusion_matrices.append(conf_matrix)

# Calculate the mean and standard deviation of accuracy scores
mean_accuracy = np.mean(accuracy_scores)
std_accuracy = np.std(accuracy_scores)

print(f"Mean Accuracy: {mean_accuracy}")
print(f"Standard Deviation of Accuracy: {std_accuracy}")

# Plot a graph showing the distribution of accuracy scores
plt.figure(figsize=(8, 6))
plt.hist(accuracy_scores, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Accuracy Scores')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')
plt.show()

# Display one confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(confusion_matrices[0], annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
